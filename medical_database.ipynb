```python
import re
df['medicine_desc_cleaned'] = df['medicine_desc'].apply(remove_html_tags)

```

```python
df['cleaned_text'] = df['name'] + " " + df['manufacturer_name'] + " " + df['type'] + " " + \
                     df['pack_size_label'] + " " + df['short_composition1'] + " " + \
                     df['short_composition2'] + " " + df['medicine_desc_cleaned']
```

```python
df['cleaned_text'] = df['cleaned_text'].str.lower()
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: " ".join(x.split()))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = vectorizer.fit_transform(df['cleaned_text'])
```

```python
def get_recommendations(medicine_name, df, tfidf_matrix):
    try:

        idx = df[df['name'] == medicine_name].index[0]
    except IndexError:
        return f"Medicine '{medicine_name}' not found in the dataset."


    medicine_vector = tfidf_matrix[idx]


    sim_scores = cosine_similarity(medicine_vector, tfidf_matrix)


    sim_scores = list(enumerate(sim_scores[0]))


    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    sim_scores = sim_scores[1:6]


    medicine_indices = [i[0] for i in sim_scores]


    return df['name'].iloc[medicine_indices]


recommendations = get_recommendations('Augmentin 625 Duo Tablet', df, tfidf_matrix)
print(f"Recommendations for 'Augmentin 625 Duo Tablet':\n")
print(recommendations)
```
