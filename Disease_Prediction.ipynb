```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
for col in df.columns:
    if 'Symptom' in col:
      df[col] = df[col].str.strip()
```

```python
df = df.fillna(pd.NA)
```

```python
df_long = pd.melt(df, id_vars=['Disease'], var_name='Symptom_N', value_name='Symptom')
```

```python
df_long = df_long.dropna()
df_long = df_long.drop(columns=['Symptom_N'])
```

```python
df_long.head(10)
```

```python
model_df = pd.crosstab(df_long['Disease'], df_long['Symptom'])
```

```python
model_df.head()
```

```python
model_df.to_csv('disease_symptoms_encoded.csv')
```
